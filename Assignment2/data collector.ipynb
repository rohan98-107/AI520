```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/mineCtRates.csv', index_col=0)
```

```python
df
```

```python
dims = [10, 20, 30]

for dim in dims:
    tmp = df[df['dim'] == dim]
    
    plt.figure(figsize=(12,8), dpi=500)
    plt.plot(tmp['density'].values, tmp['mineCtRate'].values)
    plt.plot(tmp['density'].values, tmp['baseRate'].values)

    plt.legend(['Brute + Lin Alg w/ Mine Count', 'Brute + Lin Alg w/o Mine Count'])
    
    plt.xlabel('Mine density')
    plt.ylabel('Mine Flag Rate')
    
    plt.title('Mine flag rate vs. density for brute with vs. without mine count, dim = {}'.format(dim))
    plt.savefig('./mineCtRates{}.png'.format(dim))
```
