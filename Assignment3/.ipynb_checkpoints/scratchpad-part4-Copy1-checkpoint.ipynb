```python
from Agent import *
from Landscape import *
from Cell import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Rule 1, 2 comparision on the same map; reset target each iter**

```python
num_trials = 50
dim = 50

test_landscape = landscape(dim)
test_landscape.saveLandscape('ruleThreeFourComparison')

num_searches = []

for n in range(num_trials):
    
    test_agent3 = agent(test_landscape, 3, start_i = 24, start_j = 24)
    test_agent4 = agent(test_landscape, 4, start_i = 24, start_j = 24)

    target1 = test_agent3.findTarget()
    #test_agent1.saveBelief('ruleOneTwoComparison_agentOneFinal')
    #print('Rule 1 agent found target at {} with {} actions'.format(target1, test_agent1.num_actions))

    target2 = test_agent4.findTarget()
    #test_agent2.saveBelief('ruleOneTwoComparison_agentTwoFinal')
    #print('Rule 2 agent found target at {} with {} actions'.format(target1, test_agent2.num_actions))
    
    num_searches.append([test_agent3.num_actions, test_agent4.num_actions])
    
    test_landscape.resetTarget()
    #print('new target at ({}, {})'.format(test_landscape.target_x, test_landscape.target_y))
    
    if (n+1) % (num_trials/10) == 0:
        print('trials', round((((n+1)/num_trials)*100), 2), 'percent complete')
    
    #print('\n\n')
```

```python
fixedMapComparisonDF = pd.DataFrame(num_searches)
fixedMapComparisonDF.columns = ['RuleThree', 'RuleFour']
fixedMapComparisonDF['diff'] = fixedMapComparisonDF['RuleThree'] - fixedMapComparisonDF['RuleFour']
```

**Rule 1, 2 comparision on the multiple maps; reset target each iter**

```python
num_maps = 20
num_trials = 10
dim = 50

num_searches = []

for m in range(num_maps):
    test_landscape = landscape(dim)
    
    ag1_agg = 0
    ag2_agg = 0
    
    for n in range(num_trials):  
        
        test_agent1 = agent(test_landscape, 3, start_i = 24, start_j = 24)
        test_agent2 = agent(test_landscape, 4, start_i = 24, start_j = 24)
        
        
        target1 = test_agent1.findTarget()
        target2 = test_agent2.findTarget()

        ag1_agg += test_agent1.num_actions
        ag2_agg += test_agent2.num_actions
        
        test_landscape.resetTarget()
        
        '''
        if n % (num_trials/10) == 0:
            print('\ttrials', round(((n/num_trials)*100), 2), 'percent complete')
        '''
        
    num_searches.append([(ag1_agg/num_trials), (ag2_agg/num_trials)])

    if (m+1) % (num_maps/5) == 0:
        print('maps', round(((m+1)/num_maps)*100, 2), 'percent done')
```

```python
multimap_df = pd.DataFrame(num_searches)
```

```python
multimap_df.columns=['RuleThree', 'RuleFour']
```

```python
multimap_df['diff'] = multimap_df['RuleThree'] - multimap_df['RuleFour']
```

```python
multimap_df.to_csv('./data/multimap_comparison_ruleThreeFour.csv')
```

```python
plt.figure(figsize=(12,8), dpi=500)
plt.scatter(multimap_df.index, multimap_df['RuleThree'])
plt.scatter(multimap_df.index, multimap_df['RuleFour'])
plt.legend(['Rule Three', 'Rule Four'])
plt.xlabel('Map number')
plt.ylabel('Number of searches to find target (avg over 30 trials)')
plt.title('Number of searches to find target, Rule Three vs. Rule Four, for multiple maps dim = 50')
plt.savefig('./imgs/multi_map_comparison_ruleThreeFourScatter.png')
```

```python
plt.figure(figsize=(12,8), dpi=500)
plt.plot(multimap_df.index, multimap_df['diff'])
plt.xlabel('Map number')
plt.ylabel('Diff. in number of searches to find target (avg over 30 trials)')
plt.title('Difference (Rule Three - Rule Four) in number of searches to find target, for multiple maps dim = 50')
plt.savefig('./imgs/multi_map_comparison_ruleThreeFourDiff.png')
```

```python
plt.figure(figsize=(12,8), dpi=500)
plt.boxplot(multimap_df['diff'], vert=False)
plt.xlabel('Diff in. number of searches to find target (avg over 30 trials)')
plt.title('Difference (Rule Three - Rule Four) in number of searches to find target, for multiple maps dim = 50')
plt.savefig('./imgs/multi_map_comparison_ruleThreeFourDiffBox.png')
```
